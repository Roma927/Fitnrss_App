```python
def outlier_handling(x):
    x = x.clip(upper=x.quantile(0.98))
    x = x.clip(lower=x.quantile(0.02))
    return x
```

```python
outlier_cols = num_cols
data = df[outlier_cols] = df[outlier_cols].apply(outlier_handling)
data
```
